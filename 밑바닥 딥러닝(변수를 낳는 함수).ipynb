```python
import numpy as np
#variable 클래스 구현해보기
class Variable:
  def __init__(self,data):
    self.data=data
```

```python
data=np.array(1.0)
x=Variable(data)

print(x.data)

x.data=np.array(2.0)

print(x.data)
```

```python
d=np.array([[1,2,3],[4,5,6]])
y=Variable(d)
print(y.data)
```

```python
#제곱함수 클래스 구현해보기
class Function :
  def __call__(self,input):
    x=input.data
    y=x**2
    output=Variable(y)
    return output

x=Variable(np.array(10))
f=Function()
y=f(x)

print(y.data)



#원의 넓이를 구하는 클래스 구현해보기
class Circle:
  def __call__(self,input):
    x=input.data
    y=x*x*3.14
    output=Variable(y)
    return output


x=Variable(np.array(5))
c=Circle()
y=c(x)

print(y.data)

```

```python
#제곱함수 수정해보기(모든 함수가 제공하는 기능)
class Function:
  def __call__(self,input):
    x=input.data
    y=self.forward(x)
    output=Variable(y)
    return output
  def forward(self,x):
    raise NotImplementedError()

#새로만든 제곱함수 
class Square(Function):
  def forward(self,x):
    return x**2

```
